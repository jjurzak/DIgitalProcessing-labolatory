```python
import numpy as np
import matplotlib.pyplot as plt

def idft_matrix(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    W = np.exp(2j * np.pi * k * n / N) / N
    return W @ x

x_mu = np.array([6, 2, 4, 3, 4, 5, 0, 0], dtype=complex)

time_signal = idft_matrix(x_mu)

plt.figure(figsize=(10, 5))

plt.stem(np.real(time_signal), markerfmt='C0o', basefmt='C0:', linefmt='C0:', label='Real Part')
plt.stem(np.imag(time_signal), markerfmt='C1o', basefmt='C1:', linefmt='C1:', label='Imaginary Part')

plt.plot(np.real(time_signal), 'C0-', lw=0.5)
plt.plot(np.imag(time_signal), 'C1-', lw=0.5)

plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Synthesized Signal using IDFT - variant 1')
plt.legend()
plt.grid()
plt.show()
```
